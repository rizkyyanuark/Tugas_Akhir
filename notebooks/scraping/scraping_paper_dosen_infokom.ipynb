```python
import json
import time
import requests
from groq import Groq

# Configuration
GROQ_API_KEY = "***REMOVED***"
client = Groq(api_key=GROQ_API_KEY)

print("="*100)
print("🤖 MEMULAI AGEN AI (TOOL USE) MENCARI PAPER...")
print("="*100)

# ==========================================
# 1. DEFINISIKAN ALAT (TOOL) DI DUNIA NYATA
# ==========================================
def search_semantic_scholar_by_title(title: str) -> str:
    """Fungsi Python Asli yang akan dijalankan otmatis oleh Groq AI"""
    print(f"   [🛠️ AI ACTION]: Menggunakan mesin pencari jurnal untuk mencari: '{title}'...")
    url = f"https://api.semanticscholar.org/graph/v1/paper/search?query={title}&limit=1&fields=title,authors,venue,publicationTypes,abstract,externalIds"
    try:
        response = requests.get(url, timeout=10)
        data = response.json()
        if data.get("data"):
            # Karena ini simulasi, kita kembalikan JSON mentah agar AI yang membacanya
            return json.dumps(data["data"][0])
        return "Paper tidak ditemukan di Semantic Scholar."
    except Exception as e:
        return f"Error jaringan saat mencari paper: {e}"

# ==========================================
# 2. DAFTARKAN ALAT TERSEBUT KE DALAM OTAK AI (GROQ)
# ==========================================
# Ini seperti memberi "Buku Petunjuk" ke LLM
tools = [
    {
        "type": "function",
        "function": {
            "name": "search_semantic_scholar_by_title",
            "description": "Cari data jurnal akademik asli di internet berdasarkan judul aslinya. Me-return Abstrak, Penulis, DOI, dan Jurnal tempatnya diterbitkan.",
            "parameters": {
                "type": "object",
                "properties": {
                    "title": {
                        "type": "string",
                        "description": "Judul spesifik dari paper/jurnal yang ingin dicari.",
                    }
                },
                "required": ["title"],
            },
        },
    }
]

# ==========================================
# 3. PROSES AGEN AI (TOOL-USE LOOP)
# ==========================================
SYSTEM_PROMPT = """You are an autonomous Academic Data Assistant. 
You are given a paper title by the user. 
STEPS:
1. Use your tools to search the internet for the exact abstract and metadata of the paper.
2. After receiving data from the tools, construct a COMPLETE AND EXACT formatted string block with 9 sections just like the user's ETL dashboard layout (Title, Authors, Author IDs, Journal, Document Type, Abstract, Keywords, DOI, TLDR).
3. The TLDR must be exactly 2 sentences explicitly covering the academic ontology (Problem, Task, Field, Method, Model, Dataset, Metric).
4. The Keywords must be 4-5 words separated by commas.
5. If the Author ID is missing, write 'None'.
6. Do NOT write any conversational prefix or suffix. Just return the extraction block with the '=' borders."""

# Pertanyaan pengguna
user_query = "Tolong cari dan formatkan paper ini secara lengkap: 'Applying Data Mining For Anomalies Detection On The Academic Performance Of Student'"

messages = [
    {"role": "system", "content": SYSTEM_PROMPT},
    {"role": "user", "content": user_query}
]

# Tahap 1: Tanya Llama mau melakukan apa
print("   [🧠 AI THINKING]: Memikirkan langkah untuk memecahkan permintaan User...")
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile", # 70B sangat cerdas dalam Function Calling
    messages=messages,
    tools=tools,
    tool_choice="auto",
    max_tokens=2048
)

response_message = response.choices[0].message
tool_calls = response_message.tool_calls

# Tahap 2: Jika AI memutuskan pakai alat (Tools), maka eksekusi
if tool_calls:
    # Catat histori percakapan
    messages.append(response_message)
    
    for tool_call in tool_calls:
        function_name = tool_call.function.name
        function_args = json.loads(tool_call.function.arguments)
        
        # Panggil fungsi aslinya jika AI menyuruh memanggil nama tersebut
        if function_name == "search_semantic_scholar_by_title":
            function_response = search_semantic_scholar_by_title(
                title=function_args.get("title")
            )
            
            # Tahap 3: Kirim data asli internet kembali ke otak AI
            messages.append(
                {
                    "tool_call_id": tool_call.id,
                    "role": "tool",
                    "name": function_name,
                    "content": function_response,
                }
            )

    # Tahap 4: AI membaca data asli, lalu memformat ulang jadi Laporan Final
    print("   [🧠 AI THINKING]: Menemukan data! Sedang membaca abstrak dan merajut TLDR 2-Kalimat Ontologi...")
    second_response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=messages
    )
    
    final_output = second_response.choices[0].message.content
    print("\n" + final_output)
else:
    # Jika sekiranya AI bisa jawab langsung tanpa butuh ambil data internet
    print("\n" + response_message.content)

```

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from peft import PeftModel

# Quantization config
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,
)

# Load base model
base_model = AutoModelForCausalLM.from_pretrained(
    "Qwen/Qwen2.5-0.5B-Instruct",
    quantization_config=bnb_config,
    device_map="auto",
)

# Load adapter
model = PeftModel.from_pretrained(
    base_model,
    "tcy93/scitldr-qwen2.5-0.5b-summarizer"
)
model.eval()

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-0.5B-Instruct")

def summarize_expert_en(abstract: str) -> str:
    system_prompt = """You are an AI assistant specializing in academic data extraction.
Your Task: Read the following journal abstract and summarize it into EXACTLY 2 SENTENCES in ENGLISH that are very dense and specific.

ONTOLOGY SCHEMA RULES:
[Problem], [Task], [Field], [Method], [Model], [Innovation], [Dataset], [Tool], [Metric].

STRUCTURE:
Sentence 1: Background & Approach (Problem/Task, Field, Method).
Sentence 2: Experiments & Results (Dataset, Tool, Metric).

EXAMPLE:
Abstract: "Penelitian ini bertujuan mendeteksi hoaks pada Twitter menggunakan algoritma BERT. Kami menggunakan dataset ID-Hoax dan mencapai akurasi 95%."
Output: This research addresses hoax detection within the Twitter social media domain using the BERT algorithm. Experiments were conducted on the ID-Hoax dataset and achieved an accuracy of 95%.

STRICT RULES:
- NO pronouns like "this method".
- THE OUTPUT MUST BE IN ENGLISH.
- DO NOT explain what is missing."""

    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": f"ORIGINAL ABSTRACT:\n{abstract}\n\nOUTPUT 2-SENTENCE TL;DR IN ENGLISH:"},
    ]

    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    inputs = tokenizer(text, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=200,
            do_sample=True,
            temperature=0.2,
            top_p=0.9,
            repetition_penalty=1.2
        )

    response = tokenizer.decode(
        outputs[0][inputs["input_ids"].shape[1]:],
        skip_special_tokens=True
    )
    return response.strip()

# Testing with English output
abstract_input = "Perkembangan industri hiburan digital yang pesat menuntut adanya sistem yang dapat membantu pengguna menemukan konten yang yang relevan secara cepat dan akurat. Fokus utama pengembangan ini adalah mengatasi dua permasalahan utama dalam Collaborative Filtering, yaitu data spartsity dan cold start problem. Penelitian ini menggunakan pendekatan kuantitaif dengan metode eksperimen dan dilakukan evaluasi terhadap fungsionalitas performa dan akurasi sistem. Selain itu hasil uji pengguna menunjukkan bahwa sistem mudah digunakan, cepat dalam merespon, serta mampu memberikan rekomendasi yang relevan."
print(summarize_expert_en(abstract_input))
```

```python
# ============================================================
# TLDR GENERATOR — Ontology-Aware (Revised)
# Model: tcy93/scitldr-qwen2.5-0.5b-summarizer
# Author: revised from original setup
# ============================================================

import torch
import json
import logging
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from peft import PeftModel
from typing import Optional

# Setup logging
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")
logger = logging.getLogger(__name__)


# ============================================================
# 1. MODEL SETUP
# ============================================================

# Quantization config — hemat VRAM, cukup untuk inference
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,
)

logger.info("Loading base model...")
base_model = AutoModelForCausalLM.from_pretrained(
    "Qwen/Qwen2.5-0.5B-Instruct",
    quantization_config=bnb_config,
    device_map="auto",
)

logger.info("Loading LoRA adapter...")
model = PeftModel.from_pretrained(
    base_model,
    "tcy93/scitldr-qwen2.5-0.5b-summarizer"
)
model.eval()

logger.info("Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-0.5B-Instruct")

logger.info("Model ready.")


# ============================================================
# 2. SYSTEM PROMPT — Cleaned, No Ineffective Schema Tags
# ============================================================

SYSTEM_PROMPT = """You are an academic paper summarizer.
Read the abstract below and generate exactly 2 sentences in ENGLISH:

- Sentence 1: State the research problem, task/domain, and method used.
- Sentence 2: State the key results, and mention dataset or metric ONLY IF explicitly stated in the abstract.

Rules:
- Output must be in English regardless of abstract language (Indonesian, English, or mixed).
- Do NOT invent datasets, metrics, or methods not mentioned in the abstract.
- If dataset or metric is not mentioned, omit it — do not fabricate.
- Be specific and dense, avoid generic phrases like "this study shows".
- No pronouns like "this method" or "the proposed system"."""


# ============================================================
# 3. VALIDATION
# ============================================================

def validate_tldr(tldr: str) -> bool:
    """Validasi output: minimal 2 kalimat, panjang wajar, dalam Inggris."""
    if not tldr or len(tldr.strip()) < 20:
        return False
    # Minimal ada 1 titik (2 kalimat)
    sentences = [s.strip() for s in tldr.split('.') if len(s.strip()) > 5]
    if len(sentences) < 1:
        return False
    # Cek tidak diawali kata bahasa Indonesia yang umum
    indo_starters = ["penelitian", "makalah", "studi", "sistem", "metode", "pengembangan"]
    first_word = tldr.strip().split()[0].lower()
    if first_word in indo_starters:
        return False
    return True


# ============================================================
# 4. CORE FUNCTION
# ============================================================

def generate_tldr(
    abstract: str,
    max_new_tokens: int = 100,
    repetition_penalty: float = 1.2,
    verbose: bool = False
) -> Optional[str]:
    """
    Generate ontology-aware 2-sentence TLDR dari abstract paper.
    
    Args:
        abstract: Abstract paper (bisa Indo, Inggris, atau campur)
        max_new_tokens: Batas token output (100 cukup untuk 2 kalimat)
        repetition_penalty: Hindari repetisi token
        verbose: Print debug info
    
    Returns:
        TLDR string dalam Inggris, atau None jika validasi gagal
    """
    if not abstract or len(abstract.strip()) < 10:
        logger.warning("Abstract terlalu pendek atau kosong.")
        return None

    messages = [
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user", "content": f"Abstract:\n{abstract.strip()}\n\nTLDR:"},
    ]

    # Tokenize dengan chat template
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    inputs = tokenizer(text, return_tensors="pt").to(model.device)

    if verbose:
        logger.info(f"Input token count: {inputs['input_ids'].shape[1]}")

    # Generate — greedy decoding untuk reproducibility di pipeline KG
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,            # Greedy: deterministik, cocok untuk pipeline
            repetition_penalty=repetition_penalty,
        )

    # Decode hanya bagian output baru (bukan prompt)
    response = tokenizer.decode(
        outputs[0][inputs["input_ids"].shape[1]:],
        skip_special_tokens=True
    ).strip()

    if verbose:
        logger.info(f"Raw output: {response}")

    # Validasi
    if not validate_tldr(response):
        logger.warning(f"Validasi gagal untuk output: '{response[:80]}...'")
        return None

    return response


# ============================================================
# 5. BATCH PROCESSING — Untuk Pipeline KG (Banyak Paper)
# ============================================================

def batch_generate_tldr(
    abstracts: list[dict],
    id_field: str = "paper_id",
    abstract_field: str = "abstract",
    verbose: bool = False
) -> list[dict]:
    """
    Proses banyak paper sekaligus. Input list of dict.
    
    Args:
        abstracts: List of dict, tiap dict minimal punya id_field dan abstract_field
        id_field: Nama key untuk ID paper
        abstract_field: Nama key untuk abstract
        verbose: Print progress
    
    Returns:
        List of dict dengan tambahan field 'tldr' dan 'tldr_status'
    """
    results = []
    total = len(abstracts)

    for i, paper in enumerate(abstracts):
        paper_id = paper.get(id_field, f"paper_{i}")
        abstract = paper.get(abstract_field, "")

        if verbose or i % 10 == 0:
            logger.info(f"Processing [{i+1}/{total}] paper_id: {paper_id}")

        tldr = generate_tldr(abstract, verbose=verbose)

        result = {
            **paper,  # Pertahankan semua field asli (title, author, dll.)
            "tldr": tldr,
            "tldr_status": "success" if tldr else "failed",
        }
        results.append(result)

    success_count = sum(1 for r in results if r["tldr_status"] == "success")
    logger.info(f"Batch selesai: {success_count}/{total} berhasil.")
    return results


# ============================================================
# 6. TESTING
# ============================================================

if __name__ == "__main__":

    # --- Test 1: Abstract Bahasa Indonesia ---
    print("\n" + "="*60)
    print("TEST 1: Abstract Bahasa Indonesia")
    print("="*60)

    abstract_indo = (
        "Perkembangan industri hiburan digital yang pesat menuntut adanya sistem yang dapat "
        "membantu pengguna menemukan konten yang relevan secara cepat dan akurat. "
        "Fokus utama pengembangan ini adalah mengatasi dua permasalahan utama dalam "
        "Collaborative Filtering, yaitu data sparsity dan cold start problem. "
        "Penelitian ini menggunakan pendekatan kuantitatif dengan metode eksperimen "
        "dan dilakukan evaluasi terhadap fungsionalitas performa dan akurasi sistem. "
        "Hasil uji pengguna menunjukkan bahwa sistem mudah digunakan, cepat dalam "
        "merespon, serta mampu memberikan rekomendasi yang relevan."
    )
    tldr_1 = generate_tldr(abstract_indo, verbose=True)
    print(f"Output:\n{tldr_1}\n")

    # --- Test 2: Abstract Bahasa Inggris ---
    print("="*60)
    print("TEST 2: Abstract Bahasa Inggris")
    print("="*60)

    abstract_english = (
        "We present a new approach for scientific document summarization using "
        "transformer-based models. Our method introduces title scaffolding as an "
        "auxiliary task to improve salience selection. Experiments on the SciTLDR "
        "dataset show ROUGE-1 improvements of 3.2 points over the BART baseline."
    )
    tldr_2 = generate_tldr(abstract_english, verbose=True)
    print(f"Output:\n{tldr_2}\n")

    # --- Test 3: Abstract Campur Indo-Inggris ---
    print("="*60)
    print("TEST 3: Abstract Campur Indo-Inggris")
    print("="*60)

    abstract_mixed = (
        "This paper proposes penggunaan Graph Neural Network untuk knowledge graph completion. "
        "Kami menggunakan dataset FB15k-237 dan WN18RR sebagai benchmark. "
        "Model yang diusulkan mencapai MRR 0.38 pada FB15k-237, "
        "melampaui baseline TransE dan RotatE."
    )
    tldr_3 = generate_tldr(abstract_mixed, verbose=True)
    print(f"Output:\n{tldr_3}\n")

    # --- Test 4: Batch Processing ---
    print("="*60)
    print("TEST 4: Batch Processing")
    print("="*60)

    sample_papers = [
        {
            "paper_id": "paper_001",
            "title": "Sistem Rekomendasi Berbasis Collaborative Filtering",
            "abstract": abstract_indo,
        },
        {
            "paper_id": "paper_002",
            "title": "Scientific Summarization with BART",
            "abstract": abstract_english,
        },
        {
            "paper_id": "paper_003",
            "title": "GNN for Knowledge Graph Completion",
            "abstract": abstract_mixed,
        },
    ]

    batch_results = batch_generate_tldr(
        sample_papers,
        id_field="paper_id",
        abstract_field="abstract",
        verbose=False
    )

    for res in batch_results:
        print(f"\n[{res['paper_id']}] {res['title']}")
        print(f"  Status : {res['tldr_status']}")
        print(f"  TLDR   : {res['tldr']}")

    # --- Simpan hasil ke JSON (siap untuk pipeline KG berikutnya) ---
    output_path = "tldr_results.json"
    with open(output_path, "w", encoding="utf-8") as f:
        json.dump(batch_results, f, ensure_ascii=False, indent=2)
    print(f"\nHasil disimpan ke: {output_path}")
```
