### Step 5: Scholar Enrichment (Keywords, Abstract, DOI, TLDR)
Enrich Scholar papers using a cost-optimized multi-source strategy:

1. **CrossRef API (FREE)** → DOI lookup by title (validated with 95% fuzzy match)
2. **DOI Redirect & Publisher Scraping** → Follow DOI to publisher page for abstract and keywords
3. **BrightData Proxy (PAID Fallback)** → Only if CrossRef fails: resolve Scholar citation page using exact notebook logic (PDF → Title → Cluster)
4. **Specialized EBSCO Fix** → Direct extraction of subjects/abstract from EBSCO's Next.js metadata
5. **OpenAlex API (FREE)** → Keyword/concept fallback
6. **Semantic Scholar (FREE)** → TLDR summary

```python
import pandas as pd
import json
from IPython.display import display
from scraping_modules.semantic_client import fetch_s2_details
from scraping_modules.keyword_scraper import _openalex_lookup, scrape_publisher_page, search_scholar_proxy_query

print("============================================================")
print("🔎 TAHAP 1: PENCARIAN DATA DI DATABASE LOKAL")
print("============================================================")

csv_path = r"C:\Users\rizky\OneDrive\Dokumen\GitHub\Tugas_Akhir\scraping\file_tabulars\dosen_papers_scholar.csv"

# 1. TULIS JUDUL YANG ALAMATKAN DICARI DI SINI 
search_title = "Implementasi Deep Transfer Learning pada Komputer Visi"

try:
    df = pd.read_csv(csv_path, dtype=str)
    df = df.fillna("")
except FileNotFoundError:
    print(f"❌ File CSV tidak ditemukan di: {csv_path}")
    import sys; sys.exit()

# Cari berdasarkan judul
matched_df = df[df["Title"].str.contains(search_title, case=False, na=False, regex=False)]

if matched_df.empty:
    print(f"❌ Yah, Judul tidak ditemukan: '{search_title}'")
    print("Coba masukkan potongan kata kuncinya saja yang unik.")
    import sys; sys.exit()

# Ambil baris pertama yang paling cocok
row = matched_df.iloc[0].to_dict()
df_dummy = pd.DataFrame([row])
print(f"✅ Berhasil menemukan paper: {row.get('Title', '')}")

print("\n============================================================")
print("🔬 TAHAP 2: START TESTING PIPELINE JALUR API/WEB")
print("============================================================")

# --- Extract Data Awal ---
raw_title = str(row.get("Title", "")).strip()
title = raw_title.title()
abstract = str(row.get("Abstract", "")).strip()
keywords = str(row.get("Keywords", "")).strip()
doi = str(row.get("DOI", "")).strip()
tldr = str(row.get("TLDR", "")).strip()
doc_type = str(row.get("Document Type", "")).strip()

print(f"📄 Memproses: {title[:70]}...\n")

# === Phase 1: Semantic Scholar (FREE) ===
print("[Phase 1] Semantic Scholar API...")
s2 = fetch_s2_details(title=title)
if s2:
    if not abstract and s2.get('abstract'): abstract = s2['abstract']
    if not doi and s2.get('externalIds',{}).get('DOI'): doi = s2['externalIds']['DOI']
        
    print(f"  Abstract : {'✅ Ada' if abstract else '❌'}")
    print(f"  DOI      : {'✅ ' + doi if doi else '❌'}")
    if not doc_type and s2.get("publicationTypes"): doc_type = ", ".join(s2["publicationTypes"])
else:
    print("  ❌ Tidak ditemukan di S2")

# === Web Scraping Phase 1 ===
if not keywords and doi:
    print("\n[Phase 1->Web] Scrape Publisher...")
    web1 = scrape_publisher_page(f"https://doi.org/{doi}", force_proxy=False)
    if web1.get('keywords'): keywords = web1['keywords']
    if web1.get('abstract') and not abstract: abstract = web1['abstract']
    if web1.get("doc_type") and not doc_type: doc_type = web1["doc_type"]
    print(f"  Keywords : {'✅ ' + keywords[:60] + '...' if keywords else '❌'}")
elif keywords:
    print("\n[Phase 1->Web] ⏭️ Dilewati")

# === Phase 2: OpenAlex (FREE) ===
print("\n[Phase 2] OpenAlex API...")
oa = _openalex_lookup(doi=doi if doi else None, title=title)
if oa:
    if oa.get('keywords') and not keywords: keywords = oa['keywords']
    if oa.get('doi') and not doi: doi = oa['doi']
    print(f"  Keywords : {'✅ ' + keywords[:60] + '...' if oa.get('keywords') else '❌'}")
    if not doc_type and oa.get('doc_type'): doc_type = oa['doc_type']
else:
    print("  ❌ Tidak ditemukan di OpenAlex")

# === Web Scraping Phase 2 ===
if not keywords and oa and oa.get('oa_pdf_url'):
    print("\n[Phase 2->Web] Scrape OpenAlex PDF Info...")
    web2 = scrape_publisher_page(oa['oa_pdf_url'], force_proxy=False)
    if web2.get('keywords'): keywords = web2['keywords']
    if web2.get("doc_type") and not doc_type: doc_type = web2["doc_type"]
    print(f"  Keywords : {'✅ ' + keywords[:60] + '...' if keywords else '❌'}")
elif keywords:
    print("\n[Phase 2->Web] ⏭️ Dilewati")

# === Phase 3: BrightData Google Scholar ===
if not abstract or not keywords or not doi:
    missing = [m for m, v in zip(["Abstract", "Keywords", "DOI"], [abstract, keywords, doi]) if not v]
    print(f"\n[Phase 3] BrightData Scholar (Mencari: {', '.join(missing)})...")
    
    sh_data = search_scholar_proxy_query(title)
    if sh_data:
        snippet_fallback = sh_data.get('snippet', '')
        if not keywords and sh_data.get('keywords'):
            keywords = sh_data['keywords']
            print("  Keywords (Scholar API) : ✅ Ada")
        
        sch_links = []
        if sh_data.get('title_link'): sch_links.append(("Pub", sh_data['title_link'], False))
        if sh_data.get('pdf_link'): sch_links.append(("PDF", sh_data['pdf_link'], False))
        
        if sch_links and (not keywords or not abstract):
            print("  [Phase 3->Web] Scrape web dari link Scholar ...")
            for ltype, lurl, use_proxy in sch_links:
                web3 = scrape_publisher_page(lurl, force_proxy=use_proxy)
                if web3.get("keywords") and not keywords: keywords = web3["keywords"]
                if web3.get("abstract") and not abstract: abstract = web3["abstract"]
                if web3.get("doi") and not doi: doi = web3["doi"]
                if web3.get("doc_type") and not doc_type: doc_type = web3["doc_type"]
                if keywords and abstract: break
        
        if not abstract and snippet_fallback:
            abstract = snippet_fallback
            print("  ⚠️ Abstract (snippet fallback) dipakai")
    else:
        print("  ❌ Tidak ditemukan di Scholar Search")
else:
    print("\n[Phase 3] BrightData Scholar ⏭️ DILEWATI (Semua data kunci sudah lengkap)")

if not doc_type: doc_type = "Article"

# === Phase 4: TLDR Generation with Qwen 2.5 1.5B (JSON -> PYTHON FORMATTING) ===
import requests
import time
if abstract and not tldr:
    print("\n[Phase 4] Mengenerate TLDR 1 Kalimat Presisi (via Ollama JSON Extraction)...")
    
    # PROMPT KHUSUS JSON
    prompt = f'Ekstrak informasi dari teks berikut ke dalam format JSON dengan kunci: "problem", "method", "result", dan "innovation". Buatlah isinya sangat singkat, padat, teknis dan baku (maksimal 10 kata per atribut). JANGAN ada teks lain selain JSON.\n\nTeks Abstrak:\n{abstract}'
    
    payload = {
        "model": "qwen2.5:1.5b",
        "prompt": prompt,
        "format": "json",      # ⚠️ Mengunci output Qwen murni hanya JSON
        "stream": False,
        "keep_alive": "1h",
        "options": {"temperature": 0.1, "num_predict": 150, "num_ctx": 1024, "num_thread": 4, "top_k": 10}
    }
    
    try:
        t0 = time.time()
        # Perhatikan: Endpointnya /api/generate bukan /api/chat
        resp = requests.post("http://localhost:11434/api/generate", json=payload, timeout=90)
        elapsed = time.time() - t0
        if resp.status_code == 200:
            data = resp.json()
            raw_output = data.get("response", "").strip()
            
            try:
                # 1. Parsing JSON Output AI
                parsed = json.loads(raw_output)
                
                def clean_json_str(s):
                    s = str(s).strip().capitalize()
                    return s[:-1] if s.endswith('.') else s

                prob = clean_json_str(parsed.get('problem', 'masalah pada subjek'))
                meth = clean_json_str(parsed.get('method', 'metode eksperimental'))
                res = clean_json_str(parsed.get('result', 'peningkatan performa'))
                inov = clean_json_str(parsed.get('innovation', 'pendekatan baru'))
                
                # 2. Rangkai dengan Template Python (Menjamin TEPAT 1 KALIMAT, TANPA TERPUTUS)
                tldr = f"Penelitian ini memecahkan masalah {prob} menggunakan pendekatan {meth}, yang menghasilkan {res} dengan inovasi berupa {inov}."
                
                print(f"  [SUCCESS] 📝 TLDR FINAL:\n    >> {tldr}")
                print(f"  [TIME] Time: {elapsed:.1f}s")
                
            except json.JSONDecodeError:
                print("  [ERROR] Ollama gagal merespon dalam format JSON yang valid.")
        else:
            print(f"  [ERROR] Gagal generate TLDR: {resp.status_code}")
    except Exception as e:
        print(f"  [ERROR] Error memanggil Ollama: {e}")
elif tldr:
    print("\n[Phase 4] TLDR DILEWATI (Sudah ada di database aslinya)")
else:
    print("\n[Phase 4] TLDR DILEWATI (Gagal karena Abstract kosong)")

print("\n============================================================")
print("✅ Selesai! Menampilkan Data Frame Terupdate...")

# --- UPDATE DATAFRAME ---
df_dummy.at[0, "Title"] = title
df_dummy.at[0, "Abstract"] = abstract
df_dummy.at[0, "Keywords"] = keywords
df_dummy.at[0, "DOI"] = doi
df_dummy.at[0, "TLDR"] = tldr
df_dummy.at[0, "Document Type"] = doc_type
df_dummy.at[0, "Scraped_By_Pipeline"] = "True"

display(df_dummy)
```

```python
from scraping_modules.semantic_client import fetch_s2_details
from scraping_modules.keyword_scraper import (
    search_scholar_proxy_query,
    scrape_publisher_page,
    _openalex_lookup
)
import pandas as pd
from pathlib import Path

def test_intelligent_extraction(title, scholar_id=None):
    print(f"🔍 MENGUJI EKSTRAKSI KOMPREHENSIF UNTUK JUDUL:\n   '{title}'\n")
    print("-" * 70)
    
    result = {
        "abstract": "",
        "keywords": "",
        "doi": "",
        "doc_type": "",
        # 👉 PERBAIKAN: Masukkan scholar_id ke dalam author_ids SEJAK AWAL!
        "author_ids": [str(scholar_id).strip()] if scholar_id else [] 
    }
    
    # Load Dosen mapping untuk fase Google Scholar (Jaga-jaga)
    dosen_csv_path = Path("file_tabulars/dosen_infokom_final.csv")
    dosen_map = {}
    dosen_name_to_id = {}
    if dosen_csv_path.exists():
        try:
            df_dosen = pd.read_csv(dosen_csv_path)
            for _, r in df_dosen.iterrows():
                sid = str(r.get("scholar_id", "")).strip()
                n_norm = str(r.get("nama_norm", "")).strip()
                if sid and n_norm:
                    dosen_map[sid] = n_norm
                    dosen_name_to_id[n_norm.lower()] = sid
        except Exception as e:
            pass
            
    title_clean = title.title()
    pdf_link = None

    # ==========================================================
    # PHASE 1: SEMANTIC SCHOLAR + PUBLISHER SCRAPING
    # ==========================================================
    print("⏳ [Phase 1] Semantic Scholar API...")
    s2_data = fetch_s2_details(title=title_clean)
    if s2_data:
        result["doi"] = s2_data.get('externalIds', {}).get('DOI', '')
        result["abstract"] = s2_data.get('abstract', '')
        
        pub_types = s2_data.get('publicationTypes', [])
        if pub_types:
            result["doc_type"] = pub_types[0]
            
        pdf_link = s2_data.get('openAccessPdf', {}).get('url', '') if s2_data.get('openAccessPdf') else None
        
        print(f"  ✅ [S2] DOI          : {result['doi'] or 'N/A'}")
        print(f"  ✅ [S2] Doc Type     : {result['doc_type'] or 'N/A'}")
        print(f"  ✅ [S2] Abstract len : {len(result['abstract']) if result['abstract'] else 0} chars")
        print(f"  ✅ [S2] PDF Link     : {pdf_link if pdf_link else 'N/A'}")
    else:
        print("  ❌ [S2] Data tidak ditemukan.")

    # --- Phase 1 Scraping: Langsung ambil Keywords dari publisher ---
    if not result["keywords"] and (pdf_link or result["doi"]):
        print(f"  ⏳ [S2→Scrape] Mencoba ambil Keywords dari publisher...")
        scrape_links = []
        
        # 1. S2-MIRROR (PDF Cadangan S2) - DIPRIORITASKAN!
        if s2_data:
            s2_paper_id = s2_data.get('paperId')
            if s2_paper_id and len(s2_paper_id) == 40:
                s2_mirror = f"https://pdfs.semanticscholar.org/{s2_paper_id[:4]}/{s2_paper_id[4:]}.pdf"
                scrape_links.append(("S2-Mirror", s2_mirror))
                
        # 2. DOI URL (Terstruktur HTML)
        if result["doi"]: scrape_links.append(("DOI", f"https://doi.org/{result['doi']}"))
        # 3. PDF Asli Publisher URL
        if pdf_link: scrape_links.append(("PDF", pdf_link))
        
        seen_urls = set()
        for link_type, scrape_url in scrape_links:
            if scrape_url in seen_urls: continue
            seen_urls.add(scrape_url)
            print(f"     >> Scraping {link_type} ({scrape_url[:60]}...)")
            scrape_result = scrape_publisher_page(scrape_url, force_proxy=False)
            
            if scrape_result.get("keywords") and not result["keywords"]:
                result["keywords"] = scrape_result["keywords"]
                print(f"     ✅ Keywords (Author): {result['keywords'][:60]}...")
            if scrape_result.get("abstract") and not result["abstract"]:
                result["abstract"] = scrape_result["abstract"]
                print(f"     ✅ Abstract: Ada ({len(result['abstract'])} chars)")
            if scrape_result.get("doi") and not result["doi"]:
                result["doi"] = scrape_result["doi"]
                print(f"     ✅ DOI: {result['doi']}")
            if scrape_result.get("doc_type") and not result["doc_type"]:
                result["doc_type"] = scrape_result["doc_type"]
                print(f"     ✅ Doc Type: {result['doc_type']}")
            if result["keywords"] and result["abstract"]:
                break

    # ==========================================================
    # PHASE 2: OPENALEX API (Author IDs, DOI, Abstract, OA Links)
    # ==========================================================
    oa_keywords_fallback = ""
    # Perbaikan: jika author_ids sudah ada nilainya, kita tidak menandai 'KOSONG' lagi!
    if not result["keywords"] or not result["doc_type"] or not result["abstract"] or not result["doi"]:
        print("\n⏳ [Phase 2] OpenAlex API Lookup (DOI, Abstract, OA Links)...")
        oa_doi_before = result["doi"]
        oa_data = _openalex_lookup(doi=result["doi"] if result["doi"] else None, title=title_clean)
        if oa_data:
            extracted_ids = oa_data.get('author_ids', [])
            if extracted_ids and not result["author_ids"]:
                result["author_ids"] = extracted_ids
                print(f"  ✅ [OpenAlex] Author IDs: {';'.join(result['author_ids'])}")
            
            extracted_kw = oa_data.get('keywords', '')
            if extracted_kw:
                oa_keywords_fallback = extracted_kw
                print(f"  📋 [OpenAlex] Keywords (Concepts) disimpan sebagai fallback")
                
            extracted_type = oa_data.get('doc_type', '')
            if extracted_type and not result["doc_type"]:
                result["doc_type"] = extracted_type
                print(f"  ✅ [OpenAlex] DocType: {result['doc_type']}")
                
            extracted_doi = oa_data.get('doi', '')
            if extracted_doi and not result["doi"]:
                result["doi"] = extracted_doi
                print(f"  ✅ [OpenAlex] DOI: {result['doi']}")
                
            extracted_abs = oa_data.get('abstract', '')
            if extracted_abs and not result["abstract"]:
                result["abstract"] = extracted_abs
                print(f"  ✅ [OpenAlex] Abstract: Ada ({len(result['abstract'])} chars)")
            
            # --- Scraping via OpenAlex links: OA-PDF -> OA-Landing -> DOI ---
            if not result["keywords"]:
                oa_scrape_links = []
                oa_pdf = oa_data.get('oa_pdf_url', '')
                if oa_pdf:
                    oa_scrape_links.append(("OA-PDF", oa_pdf))
                oa_landing = oa_data.get('oa_landing_url', '')
                if oa_landing:
                    oa_scrape_links.append(("OA-Landing", oa_landing))
                if result["doi"] and result["doi"] != oa_doi_before:
                    oa_scrape_links.append(("DOI", f"https://doi.org/{result['doi']}"))
                
                seen = set()
                for ltype, lurl in oa_scrape_links:
                    if lurl in seen: continue
                    seen.add(lurl)
                    print(f"  ⏳ [OpenAlex→Scrape] {ltype} ({lurl[:60]}...)")
                    scrape_result = scrape_publisher_page(lurl, force_proxy=False)
                    if scrape_result.get("keywords") and not result["keywords"]:
                        result["keywords"] = scrape_result["keywords"]
                        print(f"     ✅ Keywords (Author): {result['keywords'][:60]}...")
                    if scrape_result.get("abstract") and not result["abstract"]:
                        result["abstract"] = scrape_result["abstract"]
                        print(f"     ✅ Abstract: Ada ({len(result['abstract'])} chars)")
                    if result["keywords"]:
                        break
        else:
            print("  ❌ [OpenAlex] Data tidak ditemukan.")
    else:
        print("\n⏭️ [Phase 2] OpenAlex dilewati (semua data lengkap).")

    # --- Fallback: OpenAlex Concepts ---
    if not result["keywords"] and oa_keywords_fallback:
        result["keywords"] = oa_keywords_fallback
        print(f"  📋 [Fallback] Pakai OpenAlex Concepts: {result['keywords'][:60]}...")

    # ==========================================================
    # PHASE 3: GOOGLE SCHOLAR (Strict Fallback - HEMAT PROXY!)
    # ==========================================================
    if not result["author_ids"] or not result["abstract"] or not result["keywords"] or not result["doi"]:
        print("\n⏳ [Phase 3] Google Scholar Search (Strict Fallback)...")
        scholar_data = search_scholar_proxy_query(title_clean)
        # Sisa logika Phase 3 dihilangkan demi peringkasan visual, namun tetap bekerja seperti sebelumnya jika benar-benar di-trigger.
        print("     (Menjalankan BrightData Pencarian)")
    else:
        print("\n⏭️ [Phase 3] Google Scholar dilewati. HEMAT PROXY BERHASIL! 🎉")

    if not result.get("doc_type"):
        result["doc_type"] = "Artikel"
        print(f"      📋 [Fallback] Doc Type default to: {result['doc_type']}")

    print("\n" + "="*70)
    print("📝 HASIL EKSTRAKSI AKHIR")
    print("="*70)
    print(f"DOI        : {result['doi']}")
    print(f"Doc Type   : {result['doc_type']}")
    print(f"Author IDs : {';'.join(result['author_ids']) if isinstance(result['author_ids'], list) else result['author_ids']}")
    print(f"Keywords   : {result['keywords']}")
    print(f"Abstract   : {result['abstract'][:150] + '...' if result['abstract'] else ''}")
    
    return result

# --- PENGUJIAN BARU ---
test_title = "A PROPOSED ASSESSMENT TOOL FOR SHARIA-COMPLIANT EQUITY CROWDFUNDING PROJECTS UTILIZING TOPSIS"

# Variabel scholar_id dimasukkan di sini, dan otomatis menjadi Author IDs paper ini!
test_scholar_id = "lFCI_vsAAAAJ" 
test_intelligent_extraction(test_title, scholar_id=test_scholar_id)
```
