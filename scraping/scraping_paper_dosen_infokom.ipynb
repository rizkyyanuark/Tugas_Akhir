### Step 5: Scholar Enrichment (Keywords, Abstract, DOI, TLDR)
Enrich Scholar papers using a cost-optimized multi-source strategy:

1. **CrossRef API (FREE)** → DOI lookup by title (validated with 95% fuzzy match)
2. **DOI Redirect & Publisher Scraping** → Follow DOI to publisher page for abstract and keywords
3. **BrightData Proxy (PAID Fallback)** → Only if CrossRef fails: resolve Scholar citation page using exact notebook logic (PDF → Title → Cluster)
4. **Specialized EBSCO Fix** → Direct extraction of subjects/abstract from EBSCO's Next.js metadata
5. **OpenAlex API (FREE)** → Keyword/concept fallback
6. **Semantic Scholar (FREE)** → TLDR summary

Input: `dosen_papers_scholar_raw.csv` → Output: `dosen_papers_scholar.csv`

```python
paper_pipeline.run_scholar_enrichment()
```

```python
from scraping_modules.semantic_client import fetch_s2_details
from scraping_modules.keyword_scraper import (
    search_scholar_proxy_query,
    scrape_publisher_page,
    _openalex_lookup
)
import pandas as pd
from pathlib import Path

def test_intelligent_extraction(title):
    print(f"\U0001f50d MENGUJI EKSTRAKSI KOMPREHENSIF UNTUK JUDUL:\n   '{title}'\n")
    print("-" * 70)
    
    result = {
        "abstract": "",
        "keywords": "",
        "doi": "",
        "doc_type": "",
        "author_ids": []
    }
    
    # Load Dosen mapping untuk fase Google Scholar
    dosen_csv_path = Path("file_tabulars/dosen_infokom_final.csv")
    dosen_map = {}
    if dosen_csv_path.exists():
        try:
            df_dosen = pd.read_csv(dosen_csv_path)
            for _, r in df_dosen.iterrows():
                sid = str(r.get("scholar_id", "")).strip()
                n_norm = str(r.get("nama_norm", "")).strip()
                if sid and n_norm:
                    dosen_map[sid] = n_norm
        except Exception as e:
            print(f"   \u26a0\ufe0f Could not load dosen CSV: {e}")
            
    title_clean = title.title()
    pdf_link = None

    # ==========================================================
    # PHASE 1: SEMANTIC SCHOLAR + PUBLISHER SCRAPING
    # ==========================================================
    print("\u23f3 [Phase 1] Semantic Scholar API...")
    s2_data = fetch_s2_details(title=title_clean)
    if s2_data:
        result["doi"] = s2_data.get('externalIds', {}).get('DOI', '')
        result["abstract"] = s2_data.get('abstract', '')
        
        pub_types = s2_data.get('publicationTypes', [])
        if pub_types:
            result["doc_type"] = pub_types[0]
            
        pdf_link = s2_data.get('openAccessPdf', {}).get('url', '') if s2_data.get('openAccessPdf') else None
        
        print(f"  \u2705 [S2] DOI          : {result['doi'] or 'N/A'}")
        print(f"  \u2705 [S2] Doc Type     : {result['doc_type'] or 'N/A'}")
        print(f"  \u2705 [S2] Abstract len : {len(result['abstract']) if result['abstract'] else 0} chars")
        print(f"  \u2705 [S2] PDF Link     : {pdf_link if pdf_link else 'N/A'}")
    else:
        print("  \u274c [S2] Data tidak ditemukan.")

    # --- Phase 1 Scraping: Langsung ambil Keywords dari publisher ---
    if not result["keywords"] and (pdf_link or result["doi"]):
        print(f"  \u23f3 [S2\u2192Scrape] Mencoba ambil Keywords dari publisher...")
        scrape_links = []
        # DOI dulu (publisher HTML = terstruktur), PDF terakhir (2-kolom campur teks)
        # DOI dulu (publisher HTML = terstruktur), PDF terakhir (2-kolom campur teks)
        if result["doi"]: scrape_links.append(("DOI", f"https://doi.org/{result['doi']}"))
        if pdf_link: scrape_links.append(("PDF", pdf_link))
                
        seen_urls = set()
        for link_type, scrape_url in scrape_links:
            if scrape_url in seen_urls: continue
            seen_urls.add(scrape_url)
            print(f"     >> Scraping {link_type} ({scrape_url[:60]}...)")
            scrape_result = scrape_publisher_page(scrape_url, force_proxy=False)
            
            if scrape_result.get("keywords") and not result["keywords"]:
                result["keywords"] = scrape_result["keywords"]
                print(f"     \u2705 Keywords (Author): {result['keywords'][:60]}...")
            if scrape_result.get("abstract") and not result["abstract"]:
                result["abstract"] = scrape_result["abstract"]
                print(f"     \u2705 Abstract: Ada ({len(result['abstract'])} chars)")
            if scrape_result.get("doi") and not result["doi"]:
                result["doi"] = scrape_result["doi"]
                print(f"     \u2705 DOI: {result['doi']}")
            if result["keywords"] and result["abstract"]:
                break

    # ==========================================================
    # PHASE 2: OPENALEX API (Author IDs, DOI, Abstract, OA Links)
    # ==========================================================
    oa_keywords_fallback = ""
    if not result["author_ids"] or not result["keywords"] or not result["doc_type"] or not result["abstract"] or not result["doi"]:
        print("\n\u23f3 [Phase 2] OpenAlex API Lookup (Author IDs, DOI, Abstract, OA Links)...")
        oa_doi_before = result["doi"]
        oa_data = _openalex_lookup(doi=result["doi"] if result["doi"] else None, title=title_clean)
        if oa_data:
            extracted_ids = oa_data.get('author_ids', [])
            if extracted_ids and not result["author_ids"]:
                result["author_ids"] = extracted_ids
                print(f"  \u2705 [OpenAlex] Author IDs: {';'.join(result['author_ids'])}")
            
            extracted_kw = oa_data.get('keywords', '')
            if extracted_kw:
                oa_keywords_fallback = extracted_kw
                print(f"  \U0001f4cb [OpenAlex] Keywords (Concepts) disimpan sebagai fallback")
                
            extracted_type = oa_data.get('doc_type', '')
            if extracted_type and not result["doc_type"]:
                result["doc_type"] = extracted_type
                print(f"  \u2705 [OpenAlex] DocType: {result['doc_type']}")
                
            extracted_doi = oa_data.get('doi', '')
            if extracted_doi and not result["doi"]:
                result["doi"] = extracted_doi
                print(f"  \u2705 [OpenAlex] DOI: {result['doi']}")
                
            extracted_abs = oa_data.get('abstract', '')
            if extracted_abs and not result["abstract"]:
                result["abstract"] = extracted_abs
                print(f"  \u2705 [OpenAlex] Abstract: Ada ({len(result['abstract'])} chars)")
            
            # --- Scraping via OpenAlex links: OA-PDF -> OA-Landing -> DOI ---
            if not result["keywords"]:
                oa_scrape_links = []
                oa_pdf = oa_data.get('oa_pdf_url', '')
                if oa_pdf:
                    oa_scrape_links.append(("OA-PDF", oa_pdf))
                oa_landing = oa_data.get('oa_landing_url', '')
                if oa_landing:
                    oa_scrape_links.append(("OA-Landing", oa_landing))
                if result["doi"] and result["doi"] != oa_doi_before:
                    oa_scrape_links.append(("DOI", f"https://doi.org/{result['doi']}"))
                
                seen = set()
                for ltype, lurl in oa_scrape_links:
                    if lurl in seen: continue
                    seen.add(lurl)
                    print(f"  \u23f3 [OpenAlex\u2192Scrape] {ltype} ({lurl[:60]}...)")
                    scrape_result = scrape_publisher_page(lurl, force_proxy=False)
                    if scrape_result.get("keywords") and not result["keywords"]:
                        result["keywords"] = scrape_result["keywords"]
                        print(f"     \u2705 Keywords (Author): {result['keywords'][:60]}...")
                    if scrape_result.get("abstract") and not result["abstract"]:
                        result["abstract"] = scrape_result["abstract"]
                        print(f"     \u2705 Abstract: Ada ({len(result['abstract'])} chars)")
                    if result["keywords"]:
                        break
        else:
            print("  \u274c [OpenAlex] Data tidak ditemukan.")
    else:
        print("\n\u23ed\ufe0f [Phase 2] OpenAlex dilewati (semua data lengkap).")

    # --- Fallback: OpenAlex Concepts ---
    if not result["keywords"] and oa_keywords_fallback:
        result["keywords"] = oa_keywords_fallback
        print(f"  \U0001f4cb [Fallback] Pakai OpenAlex Concepts: {result['keywords'][:60]}...")

    # ==========================================================
    # PHASE 3: GOOGLE SCHOLAR (Strict Fallback - HEMAT PROXY!)
    # ==========================================================
    if not result["author_ids"] or not result["abstract"] or not result["keywords"] or not result["doi"]:
        print("\n\u23f3 [Phase 3] Google Scholar Search (Strict Fallback)...")
        scholar_data = search_scholar_proxy_query(title_clean)
        if scholar_data:
            extracted_ids = scholar_data.get("author_ids", [])
            if extracted_ids and not result["author_ids"]:
                mapped_ids = []
                for eid in extracted_ids:
                    if eid in dosen_map:
                        mapped_ids.append(f"{dosen_map[eid]} ({eid})")
                if mapped_ids:
                    result["author_ids"] = mapped_ids
                    print(f"  \u2705 [Scholar] Mapped Author IDs: {';'.join(result['author_ids'])}")
                else:
                    print(f"  \u274c [Scholar] Author IDs tidak match di dosen CSV.")
            
            # Simpan snippet sebagai fallback (jangan set abstract dulu)
            snippet_fallback = scholar_data.get('snippet', '')
            
            curr_kw = scholar_data.get('keywords', '')
            if not result["keywords"] and curr_kw:
                result["keywords"] = curr_kw
                print(f"  \u2705 [Scholar] Keywords: {result['keywords'][:60]}...")
            
            # --- Scrape: Publisher & PDF dulu (aman), Cache terakhir (pakai proxy) ---
            sch_links = []
            if scholar_data.get('title_link'):
                sch_links.append(("Scholar-Publisher", scholar_data['title_link'], False))
            if scholar_data.get('pdf_link'):
                sch_links.append(("Scholar-PDF", scholar_data['pdf_link'], False))
            if scholar_data.get('html_direct'):
                sch_links.append(("Scholar-HTML", scholar_data['html_direct'], False))
            # Cache terakhir, pakai proxy agar IP tidak diblokir Google
            if scholar_data.get('cached_html'):
                sch_links.append(("Scholar-Cache", scholar_data['cached_html'], True))
            
            if sch_links and (not result["keywords"] or not result["abstract"]):
                seen = set()
                for ltype, lurl, use_proxy in sch_links:
                    if lurl in seen: continue
                    seen.add(lurl)
                    print(f"  \u23f3 [{ltype}] Scraping ({lurl[:55]}...)")
                    scrape_result = scrape_publisher_page(lurl, force_proxy=use_proxy)
                    if scrape_result.get("keywords") and not result["keywords"]:
                        result["keywords"] = scrape_result["keywords"]
                        print(f"     \u2705 Keywords: {result['keywords'][:60]}...")
                    if scrape_result.get("abstract") and not result["abstract"]:
                        result["abstract"] = scrape_result["abstract"]
                        print(f"     \u2705 Abstract: Ada ({len(result['abstract'])} chars)")
                    if scrape_result.get("doi") and not result["doi"]:
                        result["doi"] = scrape_result["doi"]
                        print(f"     \u2705 DOI: {result['doi']}")
                    if result["keywords"] and result["abstract"]:
                        break
    
            # Fallback: pakai snippet jika abstract masih kosong
            if not result["abstract"] and snippet_fallback:
                result["abstract"] = snippet_fallback
                print(f"  \U0001f4cb [Scholar] Abstract (snippet fallback): {result['abstract'][:60]}...")
    else:
        print("\n\u23ed\ufe0f [Phase 3] Google Scholar dilewati. HEMAT PROXY!")

    # ==========================================================
    # HASIL KESELURUHAN
    # ==========================================================
    print("\n" + "="*70)
    print("\U0001f4dd HASIL EKSTRAKSI AKHIR")
    print("="*70)
    print(f"DOI        : {result['doi']}")
    print(f"Doc Type   : {result['doc_type']}")
    print(f"Author IDs : {';'.join(result['author_ids']) if isinstance(result['author_ids'], list) else result['author_ids']}")
    print(f"Keywords   : {result['keywords']}")
    print(f"Abstract   : {result['abstract'][:150] + '...' if result['abstract'] else ''}")
    
    return result

# --- EKSEKUSI ---
test_title = "Blended learning as a learning strategy in the disruptive era"
res = test_intelligent_extraction(test_title)
```
